# About chatGLM
With a 4096 context window, trained on 1T tokens, ChatGLM-6B is optimized for dialogue similar to ChatGPT 3.5, supplemented with RLHF (reinforcement learning with human feedback) to generate answers in-line with human preferences at just 6 Billion parameters.

With 4-bit quantization users can deploy locally on consumer GPUs (with 6GB+ of VRAM). Ask your favorite CPU inference project to add support for chatGLM today!
